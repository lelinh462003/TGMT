```python
#Import lớp Image
from PIL import Image
from imgtools import *
```

```python
# đường dẫn ảnh
dir = 'C:/Users/PhuongLinh/TGMT/01/1.jpg'
```

```python
# đọc ảnh
img = load_image(dir)
```

```python
display(img)
```

```python
# Định vị khu vực cắt
region = (150,100, 700,750)
# cắt vùng ảnh
cropped_image = img.crop(region)
display(cropped_image)
```

```python
# Dán
# xác định vị trí cần dán
paste_position = (200, 200)
img.paste(cropped_image, paste_position)
display(img)
```
